#### **Transforming Data with Python**

**1. Import library**

```python
# Import the library
import pandas as pd
```

**2. Load the data**

```python
# Load input data from the Excel file to the DataFrame
df_fbl5n = pd.read_excel('CAOT Test Case _ Final.xlsx', sheet_name = 'Input Data - FBL5N')
df_cust_hierarchy = pd.read_excel('CAOT Test Case _ Final.xlsx', sheet_name = 'Input Data - Cust Hierarchy')
df_holiday = pd.read_excel('CAOT Test Case _ Final.xlsx', sheet_name = 'Input - Holiday')
df_report_date = pd.read_excel('CAOT Test Case _ Final.xlsx', sheet_name = 'Input - Report Date', header = None, names = ['Report Date'])

# Display the input DataFrame for quick data exploration
display(df_fbl5n, df_cust_hierarchy, df_holiday, df_report_date)
```

**3. Join the DataFrame**

```python
# Select the specified columns from the df cust_hierarchy and rename them adjusted for output
df_custhi = df_cust_hierarchy[['Customer', 'Customer Name', 'TFL2', 'Customer Channel', 'SWL2', 'SWL3', 'Banner']]
df_custhi.columns = ['Customer', 'Customer Name', 'Trade Format Lv2', 'Channel', 'Sales Workforce Lv2', 'Sales Workforce Lv3', 'Banner Name']

# Perform merge and concat between df fbl5n and df cust_hierarchy
df = pd.merge(df_fbl5n.iloc[:, :2], df_custhi, how = 'left', left_on = 'Account', right_on = 'Customer')
df.drop(columns = 'Customer', inplace = True)
df = pd.concat([df, df_fbl5n.iloc[:, 2:]], axis = 1)

# Display the df DataFrame for quick overview
df.head()
```

```python
# Check the null value in the specified columns
df.isnull().sum().iloc[2:8]
```

**4. Fillna specified columns**

```python
# Fill the null value on the specified columns with 'Not Found', then check them again
df.iloc[:, 2:8] = df.iloc[:, 2:8].fillna('Not Found')
df.isnull().sum().iloc[2:8]
```

**5. Create Exclude column**

```python
# Make a function to label the Exclude column with conditional rules
def exclude(row):
    if 'CITDUMMY' in str(row['Reference Key 1']).upper():
        return row['Reference Key 1']
    elif row['Amount in local currency'] > 0:
        return 'Positive Value'
    elif row['Local Currency'] != 'IDR' or row['Document currency'] != 'IDR':
        return 'Not IDR'
    elif str(row['Document Number']).startswith('58'):
        return 'Reversal document'
    elif 'BNPPAY' in str(row['Reference']).upper() or 'BNPPAY' in str(row['Text']).upper():
        return 'BNPPAY'
    else:
        return None
    
# Create a new column called Exclude and apply the exclude function
df['Exclude'] = df.apply(exclude, axis = 1)
```

```python
# Check the value_counts of Exclude column
df['Exclude'].value_counts()
```

**6. Create Days column**

```python
# Make a function to calculates the number of working days between two dates, excluding weekends and holidays
def networkdays_intl(start_date, end_date, weekends, holidays):
    dates = pd.date_range(start=start_date, end=end_date)
    weekdays = dates[~dates.dayofweek.isin(weekends)]
    workdays = weekdays[~weekdays.isin(holidays)]
    return len(workdays)

# Make a function to label the Days column with given conditions using networkdays_intl function
def days(row1, row2):
    weekends = [5, 6]
    holidays = df_holiday['Holiday'].tolist()
    if pd.isnull(row1['Clearing Document']):
        return max(0, networkdays_intl(row1['Document Date'], row2['Report Date'][0], weekends, holidays) -1)
    else:
        return max(0, networkdays_intl(row1['Document Date'], row1['Clearing date'], weekends, holidays) -1)
        
# Create a new column called Days and apply the days function
df['Days'] = df.apply(lambda row: days(row, df_report_date), axis = 1)
```

```python
# Check the value_counts of Days column
df['Days'].value_counts()
```

**7. Create Ageing column**

```python
# Make a function to label the Ageing column based on existing Excel formula
def ageing_raw(row):
    if row['Days'] <= -1:
        return '>2'
    else:
        if row['Days'] < 2:
            return '<2'
        else:
            if row['Days'] == 2:
                return '=2'
            else:
                if row['Days'] >= 3:
                    return '>2'
```

```python
# With the domain knowledge and understanding the data, we can simplify the function
def ageing(row):
    if row['Days'] < 2:
        return '<2'
    elif row['Days'] == 2:
        return '=2'
    else:
        return '>2'

# Create a new column called Ageing and apply the ageing function
df['Ageing'] = df.apply(ageing, axis = 1)
```

```python
# Check the value_counts of Ageing column
df['Ageing'].value_counts()
```

**7. Create Clearing Status column**

```python
# Make a function to label the Clearing Status column with conditional rules
def clearing_status(row):
    return 'Open' if pd.isnull(row['Clearing Document']) else 'Cleared'

# Create a new column called Clearing Status and apply the clearing_status function
df['Clearing Status'] = df.apply(clearing_status, axis = 1)
```

```python
# Check the value_counts of Clearing Status column
df['Clearing Status'].value_counts()
```
